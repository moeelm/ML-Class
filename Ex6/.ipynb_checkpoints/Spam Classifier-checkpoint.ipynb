## Support Vector Machines

```python
%matplotlib inline
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import scipy.io
import scipy.optimize
from sklearn import svm
```

```python
datafile = os.getcwd() + '/ex6data1.mat'
mat = scipy.io.loadmat(datafile)

X,y = mat['X'], mat['y']

positive = np.array([X[i] for i in range(X.shape[0]) if y[i] == 1])
negative = np.array([X[i] for i in range(X.shape[0]) if y[i] == 0])
```

```python
def plotData():
    plt.figure(figsize=(10,6))
    plt.plot(positive[:,0], positive[:,1], 'rx', label='Positive')
    plt.plot(negative[:,0], negative[:,1], 'bo', label='Negative')
    plt.legend
plotData()
    
```

### SVM Decision Boundary

```python
def plotBoundary(my_svm, xmin, xmax, ymin, ymax):
    xvals = np.linspace(xmin, xmax, 100)
    yvals = np.linspace(ymin, ymax, 100)
    zvals = np.zeros((len(xvals), len(yvals)))
    
    for i in range(len(xvals)):
        for j in range(len(yvals)):
            zvals[i][j] = float(my_svm.predict(np.array([[xvals[i],yvals[j]]])))
    zvals = zvals.transpose()
    
    u,v = np.meshgrid(xvals, yvals)
    
    mycontour = plt.contour(xvals, yvals, zvals, [0])
    plt.title('Decision Boundary')
```

```python
#Using linear SVM kernel with C = 1
linear_svm = svm.SVC(C=1, kernel = 'linear')
linear_svm.fit(X, y.flatten())

plotData()
plotBoundary(linear_svm, 0.0, 4.5,1.5,5)
```

```python
#Using SVM with linear kernel C=100
linear_svm = svm.SVC(C=100, kernel = 'linear')
linear_svm.fit(X, y.flatten())

plotData()
plotBoundary(linear_svm, 0, 4.5, 1.5, 5)
```

### Gaussian Kernel

```python
def gaussKernel(x1,x2, sigma):
    sigmasquared = np.power(sigma, 2)
    return np.exp(-(x1-x2).T.dot(x1-x2)/(2*sigmasquared))

print (gaussKernel(np.array([1,2,1]), np.array([0,4,-1]), 2.))
```

```python
def preProcess(email):
    #Process email to make easier for classification, similar to normalizing everything to be treated the same. Assists with computational time
    
    #Make entire email lower case
    email = email.lower()
    #Strip html tags and replace with a space
    email = re.sub('<[^<>]+>', ' ', email)
    #Replace any numbers with the string 'number'
    email = re.sub('[0-9]+', 'number', email)
    #Replace anything starting with http or https:// with 'httpaddr'
    email = re.sub('(http|https)://[^\s]*', 'httpaddr', email)
    #Consider strings with the symbol '@' as emails
    email = re.sub('[^\s]+@[^\s]+', 'emailaddr', email)
    #Replace '$' with dollar
    email = re.sub('[$]+', 'dollar', email)
    
    return email
```

```python
def email2TokenList(raw_email):
    #Stems preprocessed email and returns list of tokens in email
    stemmer = nltk.stem.porter.PorterStemmer()
    email = preProcess(raw_email)
    #Split email into words (split by using ' ', '@', '$', '/')
    tokens = re.split('[ \@\$\/\#\.\-\:\&\*\+\=\[\]\?\!\(\)\{\}\,\'\"\>\_\<\;\%]', email)
    #Loop over each token/word and use the stemmer to shorten it, check if word is in the vocab list, store the index in vocab_list
    tokenlist = []
    
    for token in tokens:
        #Remove non alphanumeric characters
        token = re.sub('[^a-zA-Z0-9]', '', token)
        #Use Porter stemmer to stem the word
        stemmed = stemmer.stem(token)
        #Throw out empty tokens
        if not len(token):
            continue
        #Store list of all unique stemmed words
        tokenlist.append(stemmed)
        
        return tokenlist
            
```

```python
#Vocab List
def getVocabDict(reverse=False):
    vocab_dict = {}
    with open('vocab.txt') as f:
        for line in f:
            (val, key) = line.split()
            if not reverse:
                vocab_dict[key] = int(val)
            else: vocab_dict[int(val)] = val
    return vocab_dict
```

```python
def email2VocabIndices(raw_email, vocab_dict):
    #Takes in raw email and returns list of indices corresponding to location of vocab_doct for each stemmed word in the email
    tokenlist = email2TokenList(raw_email)
    index_list = [vocab_dict[token] for token in tokenlist if token in vocab_dict]
    return index_list
```
